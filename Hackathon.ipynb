**K Fold Cross Validation**

We will add all the features to X

```python
# ==============================
# IMPORT LIBRARIES
# ==============================

import pandas as pd
import numpy as np

from sklearn.model_selection import KFold
from sklearn.metrics import r2_score

from catboost import CatBoostRegressor
```

```python

# ==============================
# LOAD DATA
# ==============================

train = pd.read_csv("/content/sample_data/train.csv")
test = pd.read_csv("/content/sample_data/test.csv")
```

```python
# ==============================
# CREATE X AND y
# ==============================

X = train.drop(["cltv", "id"], axis=1)
y = train["cltv"]
```

```python

# ==============================
# FEATURE ENGINEERING
# ==============================

# ---- Income Mapping ----

income_map = {
    "<=2L": 1,
    "2L-5L": 2,
    "5L-10L": 3,
    "More than 10L": 4
}

X["income_encoded"] = X["income"].map(income_map)


# ---- Policy Count Mapping ----

policy_map = {
    "1": 1,
    "More than 1": 2
}

X["policy_count_encoded"] = X["num_policies"].map(policy_map)


# ---- Claim Per Policy ----

X["claim_per_policy"] = (
    X["claim_amount"] /
    X["policy_count_encoded"]
)

X["claim_vintage_ratio"] = (
    X["claim_amount"] /
    (X["vintage"] + 1)
)


# ---- Income × Vintage ----

X["income_vintage"] = (
    X["income_encoded"] *
    X["vintage"]
)
```

```python
# ==============================
# CATEGORICAL COLUMNS
# ==============================

cat_cols = X.select_dtypes(include="object").columns.tolist()

print("Categorical Columns:")
print(cat_cols)

print("\nFinal Features:")
print(X.columns)


```

```python
# ==============================
# K-FOLD CROSS VALIDATION
# ==============================

kf = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scores = []
```

```python
# ==============================
# TRAINING LOOP
# ==============================

for fold, (train_idx, valid_idx) in enumerate(kf.split(X)):

    print(f"\n========== Fold {fold+1} ==========")

    # Split data

    X_train = X.iloc[train_idx]
    X_valid = X.iloc[valid_idx]

    y_train = y.iloc[train_idx]
    y_valid = y.iloc[valid_idx]


    # Create Model

    # model = CatBoostRegressor(
    #     iterations=1000,
    #     learning_rate=0.03,
    #     depth=8,
    #     loss_function='RMSE',
    #     eval_metric='R2',
    #     verbose=100
    # )
    model1 = CatBoostRegressor(
        iterations=1500,
        learning_rate=0.02,
        depth=6,
        loss_function='RMSE',
        eval_metric='R2',
        verbose=100
    )

    model2 = CatBoostRegressor(

    iterations=2000,
    learning_rate=0.015,
    depth=7,

    loss_function='RMSE',
    eval_metric='R2',

    verbose=100
)


    # Train Model

    model1.fit(
        X_train,
        y_train,
        cat_features=cat_cols,
        eval_set=(X_valid, y_valid),
        early_stopping_rounds=100
    )

    model2.fit(
        X_train,
        y_train,
        cat_features=cat_cols,
        eval_set=(X_valid, y_valid),
        early_stopping_rounds=100
    )


    # Predict

    preds1 = model1.predict(X_valid)
    preds2 = model2.predict(X_valid)

    final_preds = (preds1 + preds2) / 2
    # Calculate Score

    score = r2_score(y_valid, final_preds)

    scores.append(score)

    print(f"Fold {fold+1} R2 Score: {score}")
```

```python
# ==============================
# FINAL AVERAGE SCORE
# ==============================

avg_score = np.mean(scores)

print("\n==============================")
print("Average R2 Score:", avg_score)
print("==============================")
```

```python
# ==============================
# TRAIN FINAL MODEL 1
# ==============================

final_model1 = CatBoostRegressor(
    iterations=1500,
    learning_rate=0.02,
    depth=6,
    loss_function='RMSE',
    eval_metric='R2',
    verbose=100
)

final_model1.fit(
    X,
    y,
    cat_features=cat_cols
)

# ==============================
# TRAIN FINAL MODEL 2
# ==============================

final_model2 = CatBoostRegressor(
    iterations=2000,
    learning_rate=0.015,
    depth=7,
    loss_function='RMSE',
    eval_metric='R2',
    verbose=100
)

final_model2.fit(
    X,
    y,
    cat_features=cat_cols
)
```

```python
# ==============================
# TEST DATA FEATURE ENGINEERING
# ==============================

test_X = test.drop(["id"], axis=1)


# Income Encoding

test_X["income_encoded"] = test_X["income"].map(income_map)


# Policy Count Encoding

test_X["policy_count_encoded"] = test_X["num_policies"].map(policy_map)


# Claim Per Policy

test_X["claim_per_policy"] = (
    test_X["claim_amount"] /
    test_X["policy_count_encoded"]
)


# Income × Vintage

test_X["income_vintage"] = (
    test_X["income_encoded"] *
    test_X["vintage"]
)
```

```python
# ==============================
# FINAL PREDICTIONS
# ==============================

# Add missing feature to test_X
test_X["claim_vintage_ratio"] = (
    test_X["claim_amount"] /
    (test_X["vintage"] + 1)
)

# ==============================
# TEST PREDICTIONS
# ==============================

preds1 = final_model1.predict(test_X)

preds2 = final_model2.predict(test_X)

# Ensemble Average

final_preds = (preds1 + preds2) / 2
```

```python
# ==============================
# CREATE SUBMISSION FILE
# ==============================
submission = pd.DataFrame()

submission["id"] = test["id"]

submission["cltv"] = final_preds
```

```python
# ==============================
# SAVE SUBMISSION
# ==============================

submission.to_csv("submission.csv", index=False)

print("\nSubmission file created successfully!")
```
